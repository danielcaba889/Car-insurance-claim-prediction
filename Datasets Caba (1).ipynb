# Car insurance claims

```python
siniestros = pd.read_csv("https://raw.githubusercontent.com/danielcaba889/Car-insurance-claim-prediction/main/Car%20insurance%20claims.csv")
pd.set_option('display.max_columns', None)
siniestros.head(50)
```

```python
print(f'el dataset siniestros tiene {siniestros.shape[0] } filas y {siniestros.shape[1] } columnas')
if siniestros.shape[0] >2000 and siniestros.shape[1]>15:
  print("el dataset siniestros cumple con las condiciones del ejercicio")
else:
  print("el dataset siniestros NO cumple con las condiciones del ejercicio")
```

# Información
### El dataset cuenta con varias columnas, sin embargo podemos destacar algunas que podrían considerarse mas importantes o fundamentales.
### A la hora de hacer un analisis de siniestralidad, las compañías de seguro deben tener en cuenta distintos factores como
### -Antigüedad del vehículo (age_of_car)
### -Edad del conductor (age_of_policyholder)
### -Modelo (model)

### Por otro lado, se observa si la poliza tuvo un siniestro o no, representado por la variable 'is_claim'





# Análisis numérico de variables relevantes

```python
col_importantes_siniestros = siniestros[["age_of_car", "age_of_policyholder","is_claim"]]
col_importantes_siniestros.head()
```

```python
for variable_importante in col_importantes_siniestros:
  media = col_importantes_siniestros[variable_importante].mean()
  max = col_importantes_siniestros[variable_importante].max()
  min = col_importantes_siniestros[variable_importante].min()
  sd = col_importantes_siniestros[variable_importante].std()
  print(f'la media de  {variable_importante} es {media}')
  print(f'el max de  {variable_importante} es {max}')
  print(f'el min de  {variable_importante} es {min}')
  print(f'el desvío de  {variable_importante} es {sd}')
  print('------------------------------------------------')
```

### Data types and Null values

```python
siniestros.dtypes
```

```python
siniestros.isnull().sum()
```
